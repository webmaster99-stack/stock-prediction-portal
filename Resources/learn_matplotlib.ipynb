```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Types of Plot

### Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
plt.plot(x, y);
plt.title("Simple Line Plot");
plt.xlabel("X-axis label");
plt.ylabel("Y-axis label");
plt.grid(True)
```

### Bar Chart

```python
months = ["Jan", "Feb", "Mar", "Apr"]
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(months, revenue);
plt.title("Monthly Revenue Bar Chart");
plt.xlabel("Month");
plt.ylabel("Revenue");
```

### Histograms

```python
marks50 = np.random.randint(0, 100, 50)
marks50
```

```python
plt.hist(marks50);
plt.xlabel("Marks");
plt.ylabel("No. of students");
```

```python
plt.hist(marks50, bins=2);
```

### Scatter Plots

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 160, 155, 180, 145]
```

```python
plt.scatter(heights, weights);
plt.title("Heights vs Weights Scatter Plot");
plt.xlabel("Heights (inches)");
plt.ylabel("Weihgts (pounds)");
```

### Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, "b", label="original price");
plt.plot(days, predicted_price, "r", label="predicted price");
plt.legend();
plt.title("Plot with Legend");
plt.xlabel("Weekdays");
plt.ylabel("Price");
```

### Subplots 

```python
plt.subplot(3, 1, 1);
plt.plot(days, original_price, "b-o", label="original price");
plt.legend();
plt.title("Original Prices Over Time");
plt.xlabel("Weekdays");
plt.ylabel("Price");

plt.subplot(3, 1, 2);
plt.plot(days, predicted_price, "r-o", label="predicted price");
plt.legend();
plt.title("Predicted Prices Over Time");
plt.xlabel("Weekdays");
plt.ylabel("Price");

price_diff = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3);
plt.bar(days, price_diff, color="purple", label="Price Difference");
plt.legend();
plt.title("Difference Between Original and Predicted Prices");
plt.xlabel("Weekdays");
plt.ylabel("Price Difference");

plt.tight_layout()
```
